```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

Preprocessing Functions

```python
#Removes HTML data
def remove_html(data):
    data = data.replace('html', ' ')
    html_tag=re.compile(r'<.*?>')
    data=html_tag.sub(r' ',data)
    return data

#Removes URL data
def remove_url(data):
    url_clean= re.compile(r"https://\S+|www\.\S+")
    data=url_clean.sub(r' ',data)
    return data

#Removes Punctuations
def remove_punctuations(data):
    punct_tag=re.compile(r'[^\w\s]')
    data=punct_tag.sub(r' ',data)
    return data

#Removes Emojis
def remove_emoji(data):
    emoji_clean= re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    data=emoji_clean.sub(r' ',data)
    url_clean= re.compile(r"https://\S+|www\.\S+")
    data=url_clean.sub(r' ',data)
    return data


def white_space(data):
    return ' '.join(data.split())
```

"as a" dataset 

```python
data_path = 'dataset/as_a.csv'
df1= pd.read_csv(data_path)
df1.head()
```

```python
df1 = df1.drop(['Unnamed: 0', 'Unnamed: 0.1','issuekey', 'title','project'], axis=1)
df1.head()
```

```python
df2 = df2.drop(['issuekey', 'title'], axis=1)
df2.head()
```

"user story 2" data set

```python
data_path = 'dataset/user story 2.csv'
df3= pd.read_csv(data_path)
df3.head()
```

```python
result = pd.concat([df1, df2, df3])

# print the concatenated data frame
print(result)
```

Preprocess data set

```python
null_counts = result.isnull().sum()
# Print the null counts
print(null_counts)
```

```python
result.dropna(inplace=True)
result['storypoint'] = result['storypoint'].astype(int)
result.head()
```

```python
result['description'] = result['description'].apply(lambda x: x.lower())
result.head()
```

```python
result['description'] = result['description'].apply(lambda x: remove_html(x))
result.head()
```

```python
result['description'] = result['description'].apply(lambda x: remove_url(x))
result.head()
```

```python
result['description'] = result['description'].apply(lambda x: remove_punctuations(x))
result.head()
```

```python
result['description'] = result['description'].apply(lambda x: remove_emoji(x))
result.head()
```

```python
result['description'] = result['description'].apply(lambda x: white_space(x))
result.head()
```

```python
stop_words = set(stopwords.words('english'))
result['description'] = result['description'].apply(lambda x: " ".join([word for word in str(x).split() if word not in stop_words or word.startswith('a')]))
result.head()
```

```python
from sklearn.model_selection import train_test_split
# split the dataset into train, validation, and test sets
train_val, test = train_test_split(result, test_size=0.2, random_state=42)
train, val = train_test_split(train_val, test_size=0.25, random_state=42)

# assign the split marks
train['split_mark'] = 'train'
val['split_mark'] = 'val'
test['split_mark'] = 'test'

# concatenate the split datasets
result = pd.concat([train, val, test], ignore_index=True)
result.head()
```

```python
save_path = 'dataset/result.csv' 
result.to_csv(save_path,index=False)
```
